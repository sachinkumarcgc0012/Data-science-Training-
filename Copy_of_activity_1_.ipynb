```python
# Install Plotly
!pip install plotly -q

import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
placement = pd.DataFrame({

"Department":[
"CSE","CSE","CSE","CSE",
"AI","AI","AI",
"ECE","ECE","ECE",
"Mechanical","Mechanical","Mechanical",
"MBA","MBA","MBA"
],

"Course":[
"Data Science","Web Development","Cloud","AI",
"Machine Learning","Python","Deep Learning",
"Embedded","IoT","VLSI",
"CAD","Automation","Robotics",
"Finance","Marketing","HR"
],

"Certification":[
"AWS","Google","Azure","TensorFlow",
"TensorFlow","Python","AWS",
"Cisco","Cisco","MATLAB",
"AutoCAD","PLC","ROS",
"NISM","Google","SHRM"
],

"Company":[
"Amazon","Infosys","Microsoft","Google",
"Google","TCS","Amazon",
"Intel","Bosch","Texas",
"Tata","Siemens","ABB",
"Deloitte","Accenture","KPMG"
],

"Placed":[
80,70,60,50,
75,65,45,
55,40,35,
50,45,30,
70,60,50
]

})

placement
```

```python
fig = px.sunburst(
    placement,
    path=[
        "Department",
        "Course",
        "Certification",
        "Company"
    ],
    values="Placed",
    color="Placed",
    color_continuous_scale="Viridis",
    title="University Placement Hierarchy"
)

fig.update_layout(
    title_x=0.5,
    width=850,
    height=650
)

fig.show()
```

```python
labels = list(pd.unique(
    placement["Department"]
    .tolist()
    + placement["Certification"].tolist()
    + placement["Company"].tolist()
))

label_index = {k:v for v,k in enumerate(labels)}

source=[]
target=[]
value=[]

# Department -> Certification

for _,row in placement.iterrows():

    source.append(label_index[row["Department"]])
    target.append(label_index[row["Certification"]])
    value.append(row["Placed"])

# Certification -> Company

for _,row in placement.iterrows():

    source.append(label_index[row["Certification"]])
    target.append(label_index[row["Company"]])
    value.append(row["Placed"])
```

```python
fig = go.Figure(data=[go.Sankey(

node=dict(

pad=20,
thickness=20,
line=dict(color="black", width=0.5),
label=labels

),

link=dict(

source=source,
target=target,
value=value

)

)])

fig.update_layout(

title="Placement Flow Dashboard",
title_x=0.5,
font_size=11,
width=1000,
height=650

)

fig.show()
```
